## Modifying the Data

The data present in some columns are not necessarily fit to be used directly. This data need to be modified in order to be used

```python
movies.at[0, 'genres']
```

To change the object (dictionary) in some column into a string for easy useage

```python
import ast

def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    
    return l
```

```python
# Applying for different columns
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head()
```

Finding the director of the movies as it is also one of the major thing to be considered for recommending the movie

```python
def get_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])

    return l
    
```

```python
# removing any space present in the entries
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ', '') for i in x])
```

```python
movies.head()
```

### Changing the overview

```python
# Splitting the overview as an array
movies['overview'] = movies['overview'].apply(lambda x:x.split())

movies.at[0, 'overview']
```

## Stemming the data-set to have consistency

<ul>
<li>Stemming is the process of reducing a word to its base or root form, known as the stem. 
<li>In natural language processing (NLP), stemming is often used as a pre-processing step to normalize words and reduce them to a common form. 
<li>It helps to improve the accuracy and efficiency of various NLP tasks. For example, consider the words "jump", "jumps", "jumping", and "jumped". The resulting stemmed words would be "jump", "jump", "jump", and "jump", respectively.

```python
# Importing Natural Language Toolkit for stemming the tag entries
import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

### Function for stemming the data

```python
# Creating function to stem the values
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return ' '.join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

# Building of the Model

Since the data had been prepared and pre-processed, we can go ahead and use the data to find recommendations for any given movie.

## Finding the similarities between the movies

To find similarities between the movies, we use the CountVectorizer and the cosine_similarity.

CountVectorizer is used to convert the data set in "tags" that is as string into a vector. It essentially counts the number of times each word appears in "tags" and creates a numerical representation of the text data.

```python
# Importing Count Vectorizer to change the tags into vectors
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
cv.fit_transform(new['tags']).toarray().shape
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector[0]
```

Since we have converted the data into vector, we can find the similarity between these vectors by finding the cosine of the angle between the two vectors. This values are found and stored in the similar matrix

```python
# To find the similarities between the vectors
from sklearn.metrics.pairwise import cosine_similarity

similar = cosine_similarity(vector)
```

## Find the list of the top recommended movies

<ol>
<li> Find the top 5 values from the similar array.
<li> Return the movies at those indices

```python
# Function to find the list of recommended Movies
def recommend(movie):
    movie_index = new[new['title'] == movie].index[0]
    distances = similar[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:6]

    for i in movie_list:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```
